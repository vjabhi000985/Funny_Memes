```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
```

```python
# path of the image
PATH = r'/content/meme.jpg'

# Load the image
img = Image.open(PATH)

# Print the various details
print(f"Image Size   : {img.size}")
print(f"Image Format : {img.format}")
print(f"Image Mode   : {img.mode}")

# At last, plot the image.
plt.imshow(img)
```

```python
def insert_text(image, Text, POS):
  # Initializing ImageDraw
  DRAW = ImageDraw.Draw(image)

  # Choose a font and size
  font = ImageFont.truetype(r'/content/Inconsolata-Bold.ttf', 26)

  # Add text to the image
  DRAW.text(POS, text=Text, fill=(230, 230, 11), font=font, align="center")

  # Plot the image
  plt.imshow(image)
```

```python
# Text included
LEFT_TEXT = """Me asking for another
***deadline***"""

RIGHT_TEXT = """Kanav Sir
after giving
mutliple extension"""

# LEFT Position
LEFT_POS = (5, 50)

# RIGHT Position
RIGHT_POS = (400, 50)

# Calling the insert text fuction for left text
insert_text(img, LEFT_TEXT, LEFT_POS)

# Calling the insert text fuction for right text
insert_text(img, RIGHT_TEXT, RIGHT_POS)
```
